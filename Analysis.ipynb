### 3.1 - Convert Date and Time Columns

```python
food['Order Date and Time'] = pd.to_datetime(food['Order Date and Time'])
food['Delivery Date and Time'] = pd.to_datetime(food['Delivery Date and Time'])
```

### 3.2 - Function to Extract Discount

```python
def extract_discount(discount_str):
    """
    Extracts numeric discount values from discount description strings.

    Args:
        discount_str (str): The discount description string (e.g., "10% off", "$5 off").

    Returns:
        float: The extracted numeric discount value (percentage or fixed amount).
    """
    if 'off' in discount_str:
        # Fixed amount off
        return float(discount_str.split(' ')[0])
    elif '%' in discount_str:
        # Percentage off
        return float(discount_str.split('%')[0])
    else:
        # No discount
        return 0.0
```

We can replace NaN values with the mode of the columns

### 3.3 - NaN Values Pre-Processing

### 3.4 - Create Discount Percentage Column

```python
food["Discount Percentage"] = food["Discounts and Offers"].apply(extract_discount)
```

### 3.5 - Function to Calculate Discount Amount

```python
def calculate_discount_amount(row):
    """
    Calculates the discount amount based on order value and discount type.

    Args:
        row (pandas.Series): A row from the DataFrame containing order and discount information.

    Returns:
        float: The calculated discount amount for the order.
    """
    order_value = row["Order Value"]
    discount_perc = row["Discount Percentage"]

    if discount_perc > 1:  # Handle potential percentage values stored as decimals
        discount_amount = order_value * discount_perc / 100
    else:
        discount_amount = discount_perc

    # Handle potential NaN values (e.g., missing order values)
    return discount_amount if not np.isnan(discount_amount) else 0.0
```

### 3.6 - Apply Function to Calculate Discount Amount

```python
food["Discount Amount"] = food.apply(calculate_discount_amount, axis=1)
```

### 4.1 - Calculate Total Costs and Revenue per Order

```python
food['Revenue'] = food['Order Value'] + food['Delivery Fee'] + food['Discount Amount']
```

```python
food['Total Costs'] = food['Delivery Fee'] + food['Commission Fee'] + food['Payment Processing Fee'] - food['Refunds/Chargebacks']
```

```python
food['Profit'] = food['Revenue'] - food['Total Costs']
```

### 4.2 - Aggregate Data to Get Overall Metrics

```python
total_orders = food.shape[0]
total_revenue = food['Revenue'].sum()
total_costs = food['Total Costs'].sum()
total_profit = food['Profit'].sum()

overall_metrics = {
    "Total Orders": total_orders,
    "Total Revenue": total_revenue,
    "Total Costs": total_costs,
    "Total Profit": total_profit
}

overall_metrics
```

A positive total profit indicates that the total revenue generated from food deliveries is exceeding the total costs. This suggests that the current commission structure, delivery fees, and promotional discounts are likely working well to achieve profitability.

### 4.3 - Break Down Costs by Category

### 4.4 - Detect Potential Outliers

```python
q1 = food['Profit'].quantile(0.25)
q3 = food['Profit'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = food[(food['Profit'] < lower_bound) | (food['Profit'] > upper_bound)]

print(f"Number of potential profit outliers: {len(outliers)}")
```

### 5.1 - Filter for Profitable Orders

```python
profitable_orders = food[food['Profit'] > 0]
```

### 5.2 - Calculate the Average Commission Percentage 

```python
profitable_orders['Commission Percentage'] = (profitable_orders['Commission Fee'] / profitable_orders['Order Value']) * 100
```

### 5.3 - Calculate the Average Discount Percentage

```python
profitable_orders['Effective Discount Percentage'] = (profitable_orders['Discount Amount'] / profitable_orders['Order Value']) * 100
```

### 5.4 - Calculate the New Averages

```python
new_avg_commission_percentage = profitable_orders['Commission Percentage'].mean()
new_avg_discount_percentage = profitable_orders['Effective Discount Percentage'].mean()

print(new_avg_commission_percentage, new_avg_discount_percentage)
```

The analysis of profitable orders has yielded a new set of average commission (17.58%) and discount (16.37%) percentages that could represent an optimal balance for achieving profitability. This suggests that adjusting commission and discount structures towards these averages may improve overall financial performance.

### 5.5 - Simulate Profitability with Recommended Discounts and Commissions

```python
recommended_commission_percentage = food['Commission Fee'].mean()
recommended_discount_percentage = food['Discount Percentage'].mean()
```

### 5.6 - Calculate the Simulated Commission Fee and Discount Amount 

```python
food['Simulated Commission Fee'] = food['Order Value'] * (recommended_commission_percentage / 100)
food['Simulated Discount Amount'] = food['Order Value'] * (recommended_discount_percentage / 100)
```

### 5.7 - Recalculate Total Costs and Profit with Simulated Values

```python
food['Simulated Total Costs'] = (food['Delivery Fee'] +
                                        food['Payment Processing Fee'] +
                                        food['Simulated Discount Amount'])

food['Simulated Profit'] = (food['Simulated Commission Fee'] + food['Simulated Total Costs'] - food['Simulated Total Costs'])
```

### 5.8 - Visualize the Comparison 

```python
plt.figure(figsize=(10, 6))

# Actual profitability
sns.kdeplot(
    food["Profit"], label="Actual Profitability", fill=True, alpha=0.5, linewidth=2
)

# Simulated profitability
sns.kdeplot(
    food["Simulated Profit"],
    label="Estimated Profitability with Recommended Rates",
    fill=True,
    alpha=0.5,
    linewidth=2,
)

plt.title(
    "Comparison of Profitability in Food Delivery: Actual vs. Recommended Discounts and Commissions"
)
plt.xlabel("Profit")
plt.ylabel("Density")
plt.legend(loc="upper left")
plt.show()
```
